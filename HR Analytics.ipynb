DATA CLEANING


This involves: 
1. Removing redundant columns 
2. Renaming them for clarity 
3. Handling missing values and 
4. Addressing duplicates. 

```python
#Deleting redundant columns
redundant_columns = ["EmployeeCount", "EmployeeNumber", "Over18", "StandardHours"]
data_cleaned = data.drop(columns=redundant_columns)
```

```python
#Renaming columns
data_cleaned.rename(columns= {'Age': 'EmployeeAge'}, inplace=True)
```

```python
#Dropping duplicates
df_cleaned = data_cleaned.drop_duplicates()
```

```python
#Removing null values
data = data.dropna()
```

```python
#Cleaning individual columns (e.g., removing whitespace from string columns)
df_cleaned['JobRole'] = df_cleaned['JobRole'].str.strip()
df_cleaned['Attrition'] = df_cleaned['Attrition'].str.strip()
df_cleaned['BusinessTravel'] = df_cleaned['BusinessTravel'].str.strip()
df_cleaned['Department'] = df_cleaned['Department'].str.strip()
df_cleaned['EducationField'] = df_cleaned['EducationField'].str.strip()
df_cleaned['Gender'] = df_cleaned['Gender'].str.strip()
df_cleaned['MaritalStatus'] = df_cleaned['MaritalStatus'].str.strip()
df_cleaned['OverTime'] = df_cleaned['OverTime'].str.strip()
```

```python
#Removing Nan values
df_cleaned.dropna(inplace=True)
```
